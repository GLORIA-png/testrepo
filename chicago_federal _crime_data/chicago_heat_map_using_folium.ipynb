```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
data =pd.read_csv("/content/Chicago_Crimes_2012_to_2017.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
x=data[["Arrest","Latitude","Longitude"]]
```

```python
x
```

```python
x.isnull().sum()
```

```python
x=x.dropna()
```

```python
x.isnull().sum()
```

```python
x.shape
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

data['Arrest'] = label_encoder.fit_transform(data['Arrest'])
```

```python
x['Arrest']= label_encoder.fit_transform(x['Arrest'])
```

```python
x
```

```python
map=folium.Map(location=[41.864073,-87.706819],zoom_start=6)
```

```python
column_to_move = 'Arrest'
column = x.pop(column_to_move)
x[column_to_move] = column
```

```python
x
```

```python
HeatMap(x).add_to(map)
```

```python
map.save("outputchicago.html")
```
