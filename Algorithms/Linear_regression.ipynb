```python
#importing necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
house_sizes = np.array([800, 1000, 1200, 1500, 1800])
# Corresponding house prices in thousands of dollars
house_prices = np.array([100, 180, 230, 250, 340])
```

```python
X = house_sizes.reshape(-1, 1)

# Creating and fitting the model
model = LinearRegression()
model.fit(X, house_prices)

# Making predictions
# Let's predict the price for a new house with a size of 1300 square feet
X_new = np.array([[2000]])
predicted_price = model.predict(X_new)

# Printing prediction
print("Based on a house size of 2000 square feet, the predicted price is approximately $", predicted_price[0])
```

```python
x_new=np.linspace(700,1900,100).reshape(-1,1)
predicted_prices=model.predict(x_new)
```

```python
predicted_prices
```

```python
# Get regression coefficients
a= model.coef_[0]
b = model.intercept_

print("a:", a)
print("b:", b)
```

```python
# Print regression equation
print("Regression Equation: y = {:.2f}x + {:.2f}".format(a,b))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(house_sizes,house_prices,color='blue',label='Data Points')
plt.plot(x_new,predicted_prices,color='red',label='Regression line')
plt.xlabel('House size')
plt.ylabel('House price')
plt.title('House price prediction with linear regression')
```

```python
# Calculate R-squared
r_squared = model.score(X, house_prices)

print("R-squared (Coefficient of Determination):", r_squared)

```
