```python
import numpy as np
```

```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt
uploaded=files.upload()
```

```python
A=cv2.imread('image.jpeg')
A=cv2.cvtColor(A,cv2.COLOR_BGR2RGB)
plt.imshow(A)
```

```python
x=np.mean(A,-1);
img=plt.imshow(256-x)
img.set_cmap('gray_r')
plt.axis('off')
plt.show()

U,S,VT=np.linalg.svd(x,full_matrices=False)
S=np.diag(S)

j=0
for r in (5,20,200):
    xapprox= U[:,:r] @ S[0:r,:r] @ VT[:r,:]
    plt.figure(j+1)
    j+=1
    img=plt.imshow(256-xapprox)
    img.set_cmap('gray_r')
    plt.axis('off')
    plt.title('r=' +str(r))
    plt.show
```
