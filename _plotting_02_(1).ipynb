```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('./Car_sales.csv',parse_dates=['Latest_Launch'])
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().mean()* 100
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data.isnull().sum()
```

#### UNIVARIATE ANALYSIS

#### HISTOGRAM

```python
plt.figure(figsize=(9,6))
plt.title('Histogram')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks(rotation=45)
sns.histplot(data=data,x='Sales_in_thousands',bins=50)
plt.show()
```

```python
data.Sales_in_thousands.max()
```

```python
data.Sales_in_thousands.min()
```

#### boxplot

```python
sns.boxplot(data=data,x='Sales_in_thousands')
plt.show()
```

```python
sns.boxplot(data=data,x='Sales_in_thousands',y='Vehicle_type')
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.xticks(rotation=75)
sns.boxplot(data=data)
plt.show()
```

### kde plot

```python
sns.kdeplot(data=data.Sales_in_thousands)
plt.show()
```

```python
sns.histplot(data=data)
plt.show();
```

### line charts

```python
plt.figure(figsize=(10,6))
plt.xlabel('__year_resale_value')
plt.ylabel('price')
plt.title('graph')
sns.lineplot(data=data,x='__year_resale_value',y='Price_in_thousands')
plt.show()
```

### count plot

```python
plt.figure(figsize=(16,5))
plt.xticks(rotation=45)
sns.countplot(data=data,x='Manufacturer')
plt.show()
```

### area chart

```python
plt.figure(figsize=(18,9))
plt.xticks(rotation=45)
plt.stackplot(data.Manufacturer,data.Sales_in_thousands,labels='Sales_in_thousands')
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.xticks(rotation=90)
plt.stackplot(data.Manufacturer,data.Sales_in_thousands,data.Price_in_thousands)
plt.show()
```

```python
data.columns
```

### horrizontal barchart

```python
plt.figure(figsize=(6,12))
plt.xticks(rotation=45)
sns.barplot(data=data,x='Sales_in_thousands',y='Manufacturer')
plt.show()
```

### vertical barplot

```python
plt.figure(figsize=(12,4))
plt.xticks(rotation = 90)
sns.barplot(data=data, x='Manufacturer', y='Horsepower')
plt.show()
```

### scatter plot

```python
sns.scatterplot(data=data, x='Horsepower', y='Sales_in_thousands')
plt.show()
```

```python
sns.scatterplot(data=data, x='Horsepower', y='Sales_in_thousands', hue='Vehicle_type')
plt.show()
```

### pie chart

```python
palette_color = sns.color_palette('bright')
```

```python
plt.pie(data=data,\
x=data.Vehicle_type.value_counts(),\
explode = [0.1, 0], \
colors=palette_color,\
labels=['Passenger', 'Car'],\
autopct='%1.1f%%');
```

### Reg plot

```python
sns.regplot(data = data, x = 'Horsepower', y= 'Sales_in_thousands')
plt.show()
```

### Rel plot

```python
sns.relplot(data = data, x = 'Horsepower', y= 'Sales_in_thousands', size='Price_in_thousands')
plt.show();
```

```python
data.columns
```

### cat plot

```python
sns.catplot(data = data, x = 'Vehicle_type', y= 'Sales_in_thousands')
plt.show();
```

### Boxen plot

```python
sns.boxenplot(data=data, y='Horsepower')
plt.show()
```

### viloin plot

```python
sns.violinplot(data=data, y='Horsepower')
plt.show()
```

```python
sns.violinplot(data=data, y='Sales_in_thousands')
plt.show()
```

### joint plot

```python
sns.jointplot(data = data, x = 'Horsepower', y= 'Sales_in_thousands')
plt.show()
```

```python
sns.jointplot(data = data, x = 'Horsepower', y= 'Sales_in_thousands', kind='hex')
plt.show()
```

### swarm plot

```python
sns.swarmplot(data = data, x = 'Vehicle_type', y= 'Sales_in_thousands')
plt.show()
```
